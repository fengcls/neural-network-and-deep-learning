#### Bullet 4: Experiment with word embeddings

The model under the default setting is saved in the result folder. This set of demos is meant to explore some inner strucutres of the learned word embeddings. Many more examples can be explored. Here is just a illustration of three simple cases.

The embeddings for the days of the week are very similar. To do the contrast, the embedding for seattle is also plotted

```python
from hw4_utils import load_data
import numpy
_, _,_ , dicts = load_data(3)
dictsw2i=dicts['words2idx']
emb= numpy.load('../result/embeddings.npy')
```

```python
from matplotlib import pyplot as plt
plt.plot(emb[dictsw2i['monday'],:])
plt.plot(emb[dictsw2i['tuesday'],:])
plt.plot(emb[dictsw2i['wednesday'],:])
plt.plot(emb[dictsw2i['thursday'],:])
plt.plot(emb[dictsw2i['friday'],:])

plt.plot(emb[dictsw2i['seattle']])
plt.legend(['mon','tue','wed','thu','fri','seattle'])
```

Like the example in the homework instruction, word pairs such as F('minneapolis')-F('minnesota') and F('miami')-F('florida'). The difference is similar indicating an approximate linear representation of the city in state and state relationship.

```python
plt.plot(emb[dictsw2i['minneapolis'],:]-emb[dictsw2i['minnesota'],:])
plt.plot(emb[dictsw2i['miami'],:]-emb[dictsw2i['florida'],:])
```

We can see that with fine parameter tuning, both the shallow and deep MLP can implement the function at 8 bit. But I failed to find a parameter for 12bit input at shallow MLP that achieve 0 error. The number of neurons needed for deep MLP is less than that for shallow layer MLP. In theory for shallow MLP, the number hidden units needed grows exponentially with number of bits. And for this particular implementation, the failure to see the exponential requirement of the number of hidden units may be due to the volume of training data (1000) may not be large enough to train networks with more parameters. For the RNN implementation, it does require less hidden units, this might comes from the fact that the network take advantage richer information in the sequential structure. This set of experiments validate the advantage of deep network structure and the recursive structure.
