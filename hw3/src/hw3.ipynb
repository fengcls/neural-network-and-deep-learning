#### Bullet 3: Explain why the back-propagation algorithm works on ms-cnn

In principle, back-propagation can compute the derivative of any function. In theano, the algorithms take a computational graph and add additional nodes to the graph that provide a symbolic description of the desired derivatives. The total derivative for a computational graph considering both direct and indirect effect corresponds to this multi-stage case. Each path can be computed independently and adds up.

#### Bullet 2: Number of parameters in CNN and MLP

```python
%reset -f
```

```python
from hw3_bonus import test_para_num
test_para_num(learning_rate=0.1, n_epochs=1000, nkerns=[16, 512],L1_reg=0.00, L2_reg=0.0001,
             batch_size=128, n_hiddenLayers=2,verbose=True)
```

For CNN, Best validation score of 9.484649 % obtained at iteration 6180, with test performance 10.679649 %. It runs for 41.36m. The total number of entries in all filters for CNN is:

$16\cdot5\cdot5+512\cdot5\cdot5 =  400 + 12800 = 13200 $

For MLP, Best validation score of 81.154057 % obtained at iteration 2060, with test performance 80.414871 %. It runs for 42.34m. The total number of entries in all weight matrices for MLP is:

$(32\cdot32\cdot3)\cdot(14\cdot14\cdot16) + (14\cdot14\cdot16)\cdot(512\cdot5\cdot5) = 9633792 + 40140800 = 49774592$

CNN has a clear advantage over MLP wrt number of parameters in this case. The running time is similar showing similar computational work.
